```python
from selenium import webdriver
from time import sleep
import pandas as pd

driver = webdriver.Chrome()
driver.get('https://amazon.in')
driver.maximize_window()

print('---------------------')
print(driver.title)
print('---------------------\n')

searchInput = driver.find_element('xpath', '//*[@id="twotabsearchtextbox"]')
searchButton = driver.find_element('xpath', '//*[@id="nav-search-submit-button"]')

searchInput.send_keys('Laptops under 50000')
searchButton.click()
sleep(2)

names = []
finalPrices = []
actualPrices = []


allItems = driver.find_elements('xpath', '//div[@data-component-type="s-search-result"]')

for page in range(3):
    
    for item in allItems:
      name = item.find_element('css selector', '.a-size-medium.a-color-base.a-text-normal')
      finalPrice = item.find_element('class name', 'a-price-whole')
      actualPrice = item.find_element('css selector', '.a-price.a-text-price')

      names.append(name.text)
      finalPrices.append(finalPrice.text)
      actualPrices.append(actualPrice.text)
    
nextButton = driver.find_element('xpath','//a[text()="Next"]')
nextButton.click()
    
data = pd.DataFrame()
data['Name'] = names
data['Final Price'] = finalPrices
data['Actual Price'] = actualPrices

data.to_excel('laptops.xlsx', index=False)

driver.quit()
```
